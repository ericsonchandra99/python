```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
```

```python
# Gunakan path ke ChromeDriver (bukan Chrome.exe!)
CHROMEDRIVER_PATH = "C:/Users/a c e r/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe"  # Ubah sesuai lokasi di komputer Anda
service = Service(CHROMEDRIVER_PATH)
driver = webdriver.Chrome(service=service)

# Define URL
linkedin_url = "https://www.linkedin.com/jobs/search/?keywords=data%20scientist&location=Indonesia"
driver.maximize_window()
driver.get(linkedin_url)  # Buka halaman

# Menentukan jumlah pekerjaan yang ingin di-scrape
no_of_jobs = 1500
n_scroll = (no_of_jobs // 25) + 1
print(f"Jumlah scroll yang dibutuhkan: {n_scroll}")

# Scroll dan klik "Load More" jika ada
for i in range(n_scroll):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")  # Scroll ke bawah
    time.sleep(3)  # Tunggu halaman memuat

    try:
        # Cek apakah ada tombol "Load More"
        button = WebDriverWait(driver, 5).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Load more')]"))
        )
        button.click()
        print("Tombol 'Load More' diklik.")
        time.sleep(2)
    except:
        print("Tombol 'Load More' tidak ditemukan atau tidak tersedia.")
        continue

# Ambil daftar pekerjaan
try:
    job_list = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CLASS_NAME, "jobs-search__results-list"))
    )
    jobs = job_list.find_elements(By.TAG_NAME, "li")  # return list of job elements
    print(f"Total pekerjaan ditemukan: {len(jobs)}")
except:
    print("Gagal menemukan daftar pekerjaan.")
```

```python
job_id= []
job_title = []
company_name = []
location = []
date = []
job_link = []
for job in jobs:
    job_id0 = job.get_attribute('data-entity-urn')
    job_id.append(job_id0)

    job_title0 = job.find_element(By.CSS_SELECTOR,'h3').get_attribute('innerText')
    job_title.append(job_title0)

    company_name0 = job.find_element(By.CSS_SELECTOR,'h4').get_attribute('innerText')
    company_name.append(company_name0)

    location0 = job.find_element(By.CLASS_NAME,'job-search-card__location').get_attribute('innerText')
    location.append(location0)

    date0 = job.find_element(By.CSS_SELECTOR,'div>div>time').get_attribute('datetime')
    date.append(date0)

    job_link0 = job.find_element(By.CSS_SELECTOR,'a').get_attribute('href')
    job_link.append(job_link0)
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

jd = []
seniority = []
emp_type = []
job_func = []
industries = []

for item in range(len(jobs)):
    job_func0 = []
    industries0 = []

    # Klik job untuk melihat detail pekerjaan
    try:
        job_click_path = f'/html/body/div[1]/div/main/section[2]/ul/li[{item+1}]/a/div[1]/img'
        job_click = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, job_click_path))
        )
        job_click.click()
    except Exception as e:
        print(f"Error in clicking job: {e}")
        continue  # Skip jika tidak berhasil klik job
    
    # Tunggu sampai halaman detail job dimuat
    time.sleep(3)
    
    try:
        jd_path = '/html/body/div[1]/div/section/div[2]/div/section[1]/div/div/section'
        jd0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, jd_path))
        ).get_attribute('innerText')
    except Exception as e:
        print(f"Error in getting job description: {e}")
        jd_path = '/html/body/div[1]/div/section/div[2]/div/section[2]/div/div/section/div'
        jd0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, jd_path))
        ).get_attribute('innerText')

    is_benefit = 'Base pay range' in jd0

    if is_benefit:
        jd_path = '/html/body/div[1]/div/section/div[2]/div/section[2]/div/div/section/div'
        jd0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, jd_path))
        ).get_attribute('innerText')
        jd.append(jd0)
    else:
        jd.append(jd0)

    # Ambil seniority
    try:
        seniority_path = jd_path + '/ul/li[1]/span'
        seniority_class_name = 'description__job-criteria-text description__job-criteria-text--criteria'
        seniority0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, seniority_class_name))
        ).get_attribute('innerText')
        seniority.append(seniority0)
    except Exception as e:
        print(f"Error in getting seniority: {e}")
        seniority.append('')

    # Ambil employment type
    try:
        emp_type_path = jd_path + '/ul/li[2]/span'
        emp_type_class_name = ''
        emp_type0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, emp_type_class_name))
        ).get_attribute('innerText')
        emp_type.append(emp_type0)
    except Exception as e:
        print(f"Error in getting employment type: {e}")
        emp_type.append('')

    # Ambil job function
    try:
        job_func_path = jd_path + '/ul/li[3]/span'
        job_func_class_name = ''
        job_func0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, job_func_class_name))
        ).get_attribute('innerText')
        job_func.append(job_func0)
    except Exception as e:
        print(f"Error in getting job function: {e}")
        job_func.append('')

    # Ambil industries
    try:
        industries_path = jd_path + '/ul/li[4]/span'
        industries_class_name = ''
        industries0 = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CLASS_NAME, industries_class_name))
        ).get_attribute('innerText')
        industries.append(industries0)
    except Exception as e:
        print(f"Error in getting industries: {e}")
        industries.append('')
```
